## ReLU Activation Function

$$ReLU(x) = x^+ = \max(0, x) = \frac{x + |x| }{2} = \begin{cases} x & \text{if } x > 0 \\ 0 & \text{if } x \le 0 \end{cases} $$

- ReLU (Rectified linear unit) is the most widely used activation function in deep learning. 

- It introduces *non-linearity* into a neural network, allowing it to learn complex patterns efficiently.
    - If the input value is positive $(x>0)$, the function returns the value itself.
    - If the input value is zero or negative $(xâ‰¤0)$, the function returns $0$.

- ReLU is predominantly used in the *hidden layers* of neural networks, particularly in deep networks.
    * *Hidden Layers*: It has become the default choice for *hidden layers* due to its main advantage: it helps prevent the vanishing gradient problem. This problem occurs when gradients become extremely small during backpropagation, effectively halting the learning process in deeper layers. Because ReLU's gradient is a constant 1 for positive inputs, the error signal can flow unimpeded through the network, leading to faster and more stable training.

```python
import matplotlib.pyplot as plt
import numpy as np

# Set print options to suppress scientific notation and set a high precision. So we can see a number like 0.0000453979 instead of making it look like -> 53978687e-05.
np.set_printoptions(suppress=True, precision=10)

def relu(x):
    return np.maximum(0, x)
```

`See what sigmoid does to a tensor`

```python
input_tensor = np.linspace(-10, 10, 5) # Create tensor with 5 numbers between -10 and 10.
input_tensor
```

```python
after_relu = relu(input_tensor)
print("Tensor after ReLU:")
after_relu
```
