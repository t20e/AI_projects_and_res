### Sigmoid Activation Function

$$sigmoid(z) = \frac{1}{1 + e^{-z} }$$

* It is often called *logistic function*.

* It takes any real-valued number as input and "squashes" it into a value between 0 and 1. This makes it particularly useful for models that need to output a probability.
    - Values close to 0 become roughly 0.
    - Values close to 1 become roughly 1.
    - Values near the center (0) become approximately 0.5.

* When the model predicts a value we can round it; $value >= 0.5$ the model predicts $1$ else it predicts $0$.

* It is best suited for the *output layer* of a neural network, especially for binary classification problems.
    - *Output Layer:* In binary classification, where the goal is to predict one of two outcomes (e.g., "yes" or "no," "spam" or "not spam"), the sigmoid function is ideal. Its output can be interpreted as a probability. For example, an output of $0.85$ means there's an $85%$ probability that the input belongs to the positive class.

```python
import matplotlib.pyplot as plt
import numpy as np

# Set print options to suppress scientific notation and set a high precision. So we can see a number like 0.0000453979 instead of making it look like -> 53978687e-05.
np.set_printoptions(suppress=True, precision=10)

def sigmoid(z):
    """
    Compute the sigmoid activate function.

    Arg:
        z -- a tensor of any size.
    """
    s = 1 / (1 + np.exp(-z))
    return s
```

`See what sigmoid does to a tensor`

```python
# Create tensor with 5 values between -10 and 10.
input_tensor = np.linspace(-10, 10, 5) 
input_tensor
```

```python
after_sigmoid = sigmoid(input_tensor)
print("Tensor after sigmoid:")
after_sigmoid
```

`To make the graph look better we increase the size of the tensors`

```python
# Create tensor with 50 numbers between -5 and 5.
input_tensor = np.linspace(-5, 5, 50) 
after_sigmoid = sigmoid(input_tensor)
```
